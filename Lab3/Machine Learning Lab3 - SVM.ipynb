## Data input

```python
# importing panda library
import pandas as pd
import numpy as np

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
  
# readinag given file
# and creating dataframe
ionosphere = pd.read_csv("ionosphere.data", header = None, delimiter=",")

# adding column headings
ionosphere.columns = ['attribute 1', 'attribute 2', 'attribute 3', 'attribute 4', 'attribute 5', 
                 'attribute 6', 'attribute 7', 'attribute 8', 'attribute 9', 'attribute 10',
                 'attribute 11', 'attribute 12', 'attribute 13', 'attribute 14', 'attribute 15', 
                 'attribute 16', 'attribute 17', 'attribute 18', 'attribute 19', 'attribute 20',
                 'attribute 21', 'attribute 22', 'attribute 23', 'attribute 24', 'attribute 25', 
                 'attribute 26', 'attribute 27', 'attribute 28', 'attribute 29', 'attribute 30',
                 'attribute 31', 'attribute 32', 'attribute 33', 'attribute 34', 'good or bad', ]
  
# storing this dataframe in a csv file
ionosphere.to_csv('ionosphere.csv', index = None)

# read from csv file
ionosphere = pd.read_csv("ionosphere.csv")
```

## Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder

# Transform data format and shape so your model can process them
# Transform discrete class to numbers
labelencoder = LabelEncoder()
ionosphere['good or bad'] = labelencoder.fit_transform(ionosphere['good or bad']) # 0 is bad, 1 is good

# Shuffle the data
ionosphere = ionosphere.sample(frac=1)

# Any other transformation boosts the final performance
################## Maybe do something here ########################


########################### END ##########################

ionosphere.head()
```

```python
ionosphere_data = ionosphere.drop('good or bad', axis = 1)
ionosphere_class = ionosphere['good or bad']
```

```python
ionosphere_data.head()
```

```python
ionosphere_class[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# split to train & test
X_train,X_test,y_train,y_test=train_test_split(ionosphere_data,ionosphere_class,test_size=0.3)

# normalize train & test data
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

## Model construction: SVM(linear)

```python
from sklearn.svm import SVC

svm_linear = SVC(kernel = 'linear', probability = True)
svm_linear.fit(X_train_std, y_train)
```

### - Results

```python
# import lib
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from matplotlib import pyplot as plt
import seaborn as sns

from tabulate import tabulate

y_predict = svm_linear.predict(X_test_std)

# results: confusion matrix
mat2 = confusion_matrix(y_test, y_predict)
sns.heatmap(mat2,square= True, annot=True, cbar= False)
plt.xlabel("predicted value")
plt.ylabel("true value")
plt.show()

# results: accuracy
accuracy2 = accuracy_score(y_test, y_predict)

# results: sensitivity(recall)
recall2 = recall_score(y_test, y_predict, average='macro')

# results: precision
precision2 = precision_score(y_test, y_predict, average='macro')

# draw table with accuracy, recall, precision
data = [[accuracy2, recall2, precision2]]

headers = ["accuracy", "recall", "precision"]

print(tabulate(data, headers= headers, tablefmt= "grid"))
```
