## Train-Test-Split & Model construction: holdout validation

```python
from sklearn.model_selection import train_test_split

mushroom_data = mushroom.drop('class', axis = 1)
mushroom_data = pd.get_dummies(mushroom_data)
pd.DataFrame(mushroom_data)
mushroom_class = mushroom['class']


X_train,X_test,y_train,y_test=train_test_split(mushroom_data,mushroom_class,test_size=0.3)
```

### - Results: with laplace smoothing

```python
from sklearn.naive_bayes import MultinomialNB

mushroom_model=MultinomialNB()
mushroom_model.fit(X_train,y_train)
```

```python
# import lib
from sklearn.metrics import confusion_matrix
from sklearn.metrics.classification import accuracy_score
from sklearn.metrics.classification import recall_score
from sklearn.metrics.classification import precision_score
from tabulate import tabulate


y_predict = mushroom_model.predict(X_test)

# results: confusion matrix
mat1 = confusion_matrix(y_test, y_predict)
sns.heatmap(mat1,square= True, annot=True, cbar= False)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel("predicted value")
plt.ylabel("true value")
plt.show()

# results: accuracy
accuracy1 = accuracy_score(y_test, y_predict)

# results: sensitivity(recall)
recall1 = recall_score(y_test, y_predict, pos_label="p")

# results: precision
precision1 = precision_score(y_test, y_predict, pos_label="p")

# draw table with accuracy, recall, precision
data = [[accuracy1, recall1, precision1]]

headers = ["accuracy", "recall", "precision"]

print(tabulate(data, headers= headers, tablefmt= "grid"))
```

### - Results: without laplace smoothing

```python
mushroom_model2=MultinomialNB(alpha = 0)
mushroom_model2.fit(X_train,y_train)
```

```python
y_predict = mushroom_model2.predict(X_test)

# results: confusion matrix
mat2 = confusion_matrix(y_test, y_predict)
sns.heatmap(mat2,square= True, annot=True, cbar= False)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel("predicted value")
plt.ylabel("true value")
plt.show()

# results: accuracy
accuracy2 = accuracy_score(y_test, y_predict)

# results: sensitivity(recall)
recall2 = recall_score(y_test, y_predict, pos_label="p")

# results: precision
precision2 = precision_score(y_test, y_predict, pos_label="p")

# draw table with accuracy, recall, precision
data = [[accuracy2, recall2, precision2]]

headers = ["accuracy", "recall", "precision"]

print(tabulate(data, headers= headers, tablefmt= "grid"))
```

## Train-Test-Split & Model construction: K-fold cross-validation

```python
mushroom_model3=MultinomialNB()
mushroom_model4=MultinomialNB(alpha = 0)
```

```python
import statistics
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

m1_accscores = []
m1_recscores = []
m1_prescores = []
m2_accscores = []
m2_recscores = []
m2_prescores = []

cm1_holder = []
cm2_holder = []
    
for train_index, test_index in folds.split(mushroom_data,mushroom_class):
    X_train, X_test, y_train, y_test = mushroom_data.iloc[train_index], mushroom_data.iloc[test_index], \
                                       mushroom_class.iloc[train_index], mushroom_class.iloc[test_index]
    # test mushroom_model3
    mushroom_model3.fit(X_train, y_train)
    y_predict = mushroom_model3.predict(X_test)
    cm1_holder.append(confusion_matrix(y_test, y_predict))
    m1_accscores.append(accuracy_score(y_test, y_predict))  
    m1_recscores.append(recall_score(y_test, y_predict, pos_label="p"))  
    m1_prescores.append(recall_score(y_test, y_predict, pos_label="p"))
    
    # test mushroom_model4
    mushroom_model4.fit(X_train, y_train)
    y_predict = mushroom_model4.predict(X_test)
    cm2_holder.append(confusion_matrix(y_test, y_predict))
    m2_accscores.append(accuracy_score(y_test, y_predict))  
    m2_recscores.append(recall_score(y_test, y_predict, pos_label="p"))  
    m2_prescores.append(recall_score(y_test, y_predict, pos_label="p"))
```

### - Results: with laplace smoothing

```python
# results: confusion matrix
mat3 = sum(cm1_holder)/3
sns.heatmap(mat3,square= True, annot=True, cbar= False)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel("predicted value")
plt.ylabel("true value")
plt.show()

# results: accuracy
accuracy3 = statistics.mean(m1_accscores)

# results: sensitivity(recall)
recall3 = statistics.mean(m1_recscores)

# results: precision
precision3 = statistics.mean(m1_prescores)

# draw table with accuracy, recall, precision
data = [[accuracy3, recall3, precision3]]

headers = ["accuracy", "recall", "precision"]

print(tabulate(data, headers= headers, tablefmt= "grid"))
```

### - Results: without laplace smoothing

```python
# results: confusion matrix
mat4 = sum(cm2_holder)/3
sns.heatmap(mat4,square= True, annot=True, cbar= False)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel("predicted value")
plt.ylabel("true value")
plt.show()

# results: accuracy
accuracy4 = statistics.mean(m2_accscores)

# results: sensitivity(recall)
recall4 = statistics.mean(m2_recscores)

# results: precision
precision4 = statistics.mean(m2_prescores)

# draw table with accuracy, recall, precision
data = [[accuracy4, recall4, precision4]]

headers = ["accuracy", "recall", "precision"]

print(tabulate(data, headers= headers, tablefmt= "grid"))
```
